```python
def cipher_step(text, shift):
    result = ""
    
    for char in text:
        # if/elif/else to handle upper/lowercase and spaces(or punctuation)
        # using ord to get the ascii value of the char
        # take ascii value and sub

        # uppercase letters
        if char.isupper():
            index = ord(char) - ord('A')
            new_index = (index + shift) % 26
            result += chr(new_index + ord('A'))
        
        # lowercase letters
        elif char.islower():
            index = ord(char) - ord('a')
            new_index = (index + shift) % 26
            result += chr(new_index + ord('a'))

        # spaces and punctuation    
        else:
            result += char
            
    return result
```

```python
# Test it out
original_message = "ZPV NBZ OPU CF JOUFSFTUFE JO XBS, CVU XBS JT JOUFSFTUFE JO ZPV "
encrypted = cipher_step(original_message, 79)
decrypted = cipher_step(encrypted, -79)

print(f"Encrypted: {encrypted}")
print(f"Decrypted: {decrypted}")
```

```python
message = "CH NBY GUTY IZ BYL CGUACHUNCIH SIO VYWIGY FIMN"
decrypted = cipher_step(message, -124)
encrypted = cipher_step(decrypted, 124)

print(f"Encrypted: {encrypted}")
print(f"Decrypted: {decrypted}")
```
